```python
#Define a function, which takes one argument, a string
def bwt(dna):
    
    #add dollar signs
    dna = dna + '$'
    
    #create a new list which is then going to take the transformed string
    dna_list = []
    
    #begin a loop
    for nucleotide in range(len(dna)):
        #define a variable which saves the last letter of the string and all the letters except the last letter
        transformed_string = dna[-1] + dna[:-1]
        #define a variable to save this string (the first transformation)
        final_string = ''.join(transformed_string)
        #change the original string to become the new string
        dna = final_string
        #append the string to a list
        dna_list.append(final_string)

    return dna_list
    

bwt_two = bwt(dna_two)
```

```python
#function to put the SNPs into a dataframe
def save_df(sequences):
    
    #create new list to add to the columns
    snp_list = []
    
    #initalize loop to go through list of sequences
    for item in sequences:
        #splits SNPs into separate letters to then organize into rows
        item_one = ([*item])
        #adds SNPs to new list
        snp_list.append(item_one)
    #appends to a dataframe
    gene_df = pd.DataFrame(snp_list)
    #alphabetize dataframe
    gene_df = gene_df.sort_values(0)
    #gene_df.head()
    #get last column of df
    final_transformed = gene_df[gene_df.columns[len(gene_df.columns)-1]]
    #print(final_transformed)
    
    #take column and put it in a list
    transformed_string = final_transformed.tolist()
    
    final_string =''
    #loop through list and add it to a string
    for nucleotide in transformed_string:
        final_string +=nucleotide
        
    #return final string
    return final_string


bwt_df_two = save_df(bwt_two)
```

```python
#function to perform inverse bwt, returns dataframe which should contain original string at the first row
def ibwt(suffix_string, first_col):
    
    #initialize dataframe to have as many columns as there are in the length of the SNP
    og_df = pd.DataFrame(index = range(len(suffix_string)))
    
    #set the first column equal to the sorted bwt output
    og_df[0] = first_col
    #set the second column equal to the bwt output
    og_df[1] = suffix_string

    #do this as many times as there are characters in BWT output
    for char in suffix_string:
        
        #set an iterator which starts from the current size of the dataframe
        greater_iterator = len(og_df.columns)
        
        #make new column which equals the previous column, move each column up by one
        for i in range(greater_iterator, 0, -1):
            
            og_df[i] = og_df[i-1]
            i-=1
        
        #set first column again equal to bwt output, then sort
        og_df[0] = suffix_string
        og_df = og_df.sort_values(by=0)
        
    return og_df
```

```python
#function to turn sorted bwt output into an array
def get_first_col(suffix_string):
    
    #initalize empty array
    first_col = []

    #append chars to first col array
    for char in suffix_string:
        first_col.append(char)

    #sort columns
    first_col.sort()
    
    return first_col

#function to turn bwt output into an array
def get_suffix_array(suffix_string):
    
    #initalize array
    suffix_array = []
    
    #append chars to suffix array
    for char in suffix_string:
        suffix_array.append(char)
    
    return suffix_array

suffix_array_two = get_suffix_array(bwt_df_two)
first_col_two = get_first_col(bwt_df_two)
ibwt(suffix_array_two, first_col_two)
```

```python
#compression algorithms

#with dollar sign added
def compress_dollar(dna):
    
    #add dollar sign
    dna = dna + '$'
    
    #convert to binary
    dna_bytes = dna.encode()
    
    #compress
    compressed_bytes = bz2.compress(dna_bytes)
    
    return compressed_bytes

#without dollar sign
def compress_no_dollar(dna):
    
    #convert to binary
    dna_bytes = dna.encode()
    
    #compress
    compressed_bytes = bz2.compress(dna_bytes)
    
    return compressed_bytes

#decompression function
def decompress_data(dna):
    
    decompressed_bytes = bz2.decompress(dna)
    decompressed_data = decompressed_bytes.decode()
    
    return (decompressed_data)
```

```python
def analyze_compression():
    
    #initalize array of each SNP to test
    dna_array = [dna_one, dna_two, dna_three, dna_four, dna_five, dna_six, dna_seven, dna_eight]
    #initialize array to store compression strings, with dollar added and without
    compressed_dollars = []
    compressed_no_dollars = []
    #initialize array to store number of bytes for those with and without dollars
    bytes_dollars = []
    bytes_no_dollars = []
    #initialize array to store whether or not decompressed string matches original string
    matches_dollars = []
    matches_no_dollars = []
    
    #loop through functions to compress files, calculate bytes, decompress files and determine accuracy
    for dna in dna_array:
        
        compressed_dollar = compress_dollar(dna)
        compressed_dollars.append(compressed_dollar)
        compressed_no_dollar = compress_no_dollar(dna)
        compressed_no_dollars.append(compressed_no_dollar)
    
    #loop through functions to calculate byte size
    for file in compressed_dollars:
        
        #calculate size and then append size to array
        byte_size = sys.getsizeof(file)/8
        bytes_dollars.append(byte_size)
    
    #loop through functions to calculate byte size
    for file in compressed_no_dollars:
        
        #calculate size and then append size to array
        byte_size = sys.getsizeof(file)/8
        bytes_no_dollars.append(byte_size)
        
    
    #decompress dollars and test
    #loop through functions to determine accuracy
    for file in range(len(compressed_dollars)):
        
        #decompress files
        decompressed_dollar = decompress_data(compressed_dollars[file])
        #remove dollar sign
        decompressed_dollar = decompressed_dollar[:-1]
        #append 1 if the file matches, 0 if it doesn't match
        if (decompressed_dollar == dna_array[file]):
            matches_dollars.append(1)
        else:
            matches_dollars.append(0)
        
    #decompress no dollars and test
    #loop through functions to determine accuracy
    for file in range(len(compressed_no_dollars)):
        
        #decompress files
        decompressed_no_dollar = decompress_data(compressed_no_dollars[file])
        #append 1 if the file matches, 0 if it doesn't match
        if (decompressed_no_dollar == dna_array[file]):
            matches_no_dollars.append(1)
        else:
            matches_no_dollars.append(0)
    
    #initialize dollar and no dollar dataframes
    dollars_df = pd.DataFrame()
    no_dollars_df = pd.DataFrame()
    
    #add columns for each data point
    dollars_df['OG SNPs'] = dna_array
    no_dollars_df['OG SNPs'] = dna_array
    
    dollars_df['Compressed Dollars'] = compressed_dollars
    no_dollars_df['Compressed Without Dollars'] = compressed_no_dollars
    
    dollars_df['Bytes'] = bytes_dollars
    no_dollars_df['Bytes'] = bytes_no_dollars
    
    dollars_df['Matches File'] = matches_dollars
    no_dollars_df['Matches File'] = matches_no_dollars
    
    return dollars_df, no_dollars_df
```

```python
analyze_compression()

#graph byte size for dollars and no dollars
dollars_df,no_dollars_df = analyze_compression()
dollars_df['$'] = 'has $'
no_dollars_df['$'] = 'no $'
dss = pd.concat([dollars_df,no_dollars_df])
sns.barplot(x='OG SNPs', y='Bytes', hue='$', data=dss)
plt.xticks(rotation=45)
```
